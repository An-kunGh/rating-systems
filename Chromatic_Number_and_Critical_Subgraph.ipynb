```
import networkx as nx
from itertools import combinations

# ==========================================================
# Fungsi brute force untuk mencari chromatic number
# ==========================================================
def chromatic_number(G):
    nodes = list(G.nodes())
    n = len(nodes)

    for k in range(1, n + 1):
        colors = {}

        def valid(node, color):
            for neighbor in G.neighbors(node):
                if neighbor in colors and colors[neighbor] == color:
                    return False
            return True

        def backtrack(i):
            if i == n:
                return True

            node = nodes[i]
            for color in range(k):
                if valid(node, color):
                    colors[node] = color
                    if backtrack(i + 1):
                        return True
                    del colors[node]
            return False

        if backtrack(0):
            return k

    return n


# ==========================================================
# Mengecek apakah graf k-critical
# ==========================================================
def is_k_critical(G):
    chi_G = chromatic_number(G)

    for v in list(G.nodes()):
        H = G.copy()
        H.remove_node(v)

        if chromatic_number(H) >= chi_G:
            return False

    return True


# ==========================================================
# Cari semua subgraf induced yang critical
# ==========================================================
def find_all_critical_subgraphs(G):
    nodes = list(G.nodes())
    critical_subgraphs = []

    for r in range(1, len(nodes) + 1):
        for subset in combinations(nodes, r):
            H = G.subgraph(subset).copy()

            if len(H.nodes()) > 0 and is_k_critical(H):
                critical_subgraphs.append(H)

    return critical_subgraphs


# ==========================================================
# CONTOH PEMAKAIAN
# ==========================================================
if __name__ == "__main__":

    # contoh graf
    G = nx.Graph()

    edges = [
        (1,2), (1,3), (1,4), (1,5), (1,6),
        (2,3), (2,4), (2,9), (2,12),
        (3,6), (3,11), (3,12),
        (4,5), (4,7), (4,9),
        (5,6), (5,7), (5,8),
        (6,8), (6,11),
        (7,8), (7,9), (7,10),
        (8,10), (8,11),
        (9,10), (9,12),
        (10,11), (10,12),
        (11,12)
    ]

    G.add_edges_from(edges)

    criticals = find_all_critical_subgraphs(G)

    print("Jumlah subgraf kritis:", len(criticals))
    print()

    for i, H in enumerate(criticals, start=1):
        print(f"Subgraf kritis #{i}")
        print("Vertices:", list(H.nodes()))
        print("Edges   :", list(H.edges()))
        print("χ(H)    :", chromatic_number(H))
        print("-" * 40)
```

```
import networkx as nx

# =====================================================
# Fungsi mencari chromatic number dengan brute force
# =====================================================
def chromatic_number(G):
    nodes = list(G.nodes())
    n = len(nodes)

    # coba dari 1 warna sampai n warna
    for k in range(1, n + 1):
        colors = {}

        # cek apakah warna valid untuk node tertentu
        def valid(node, color):
            for neighbor in G.neighbors(node):
                if neighbor in colors and colors[neighbor] == color:
                    return False
            return True

        # backtracking pewarnaan
        def backtrack(i):
            if i == n:
                return True

            node = nodes[i]

            for color in range(k):
                if valid(node, color):
                    colors[node] = color

                    if backtrack(i + 1):
                        return True

                    del colors[node]

            return False

        # kalau berhasil dengan k warna
        if backtrack(0):
            return k

    return n


# =====================================================
# CONTOH PEMAKAIAN
# =====================================================
if __name__ == "__main__":

    G = nx.Graph()

    edges = [
        (1,2), (1,3), (1,4), (1,5), (1,6),
        (2,3), (2,4), (2,5), (2,6),
        (3,4), (3,5), (3,6),
        (4,5), (4,6),
        (5,6)
    ]

    G.add_edges_from(edges)

    chi = chromatic_number(G)

    print("Chromatic Number =", chi)
```
